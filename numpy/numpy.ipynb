What is NumPy?
==============

**NumPy** stands for

> Numerical Python

It is the fundamental library for numerical computing in Python.

It provides

-   Fast multidimensional arrays
-   Mathematical operations
-   Linear algebra
-   Random number generation
-   Statistics
-   Broadcasting
-   Memory-efficient storage

## 1 - What is numpy


NumPy stands for **Numerical Python**.

It is an open-source Python library designed for **fast numerical computation**.

Before NumPy existed, Python already had lists. However, Python lists were never designed for scientific computing or mathematical operations on large datasets.

NumPy solves this problem by providing a new data structure called an **ndarray (N-Dimensional Array)** that stores data much more efficiently and performs computations significantly faster.

Almost every major Data Science, Machine Learning, and AI library either directly uses NumPy arrays or is built on top of them.

Some examples include:

-   Pandas
-   Scikit-Learn
-   TensorFlow
-   PyTorch
-   SciPy
-   Matplotlib
-   OpenCV

## 2 - Why was NumPy

To understand NumPy, we first need to understand the limitations of Python itself.

Python was created as a **general-purpose programming language**.

Its priorities were:

-   Readability
-   Simplicity
-   Flexibility

Its priorities were **not**

-   High-performance mathematics
-   Scientific computing
-   Matrix operations
-   Linear algebra
-   Numerical simulations
